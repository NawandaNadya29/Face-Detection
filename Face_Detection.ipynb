```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Menginput gambar
image = cv2.imread('blekping.jpeg')

plt.imshow(image)
plt.show()
```

```python
# Filter menggunakan haarcascade Frontal face
haarcascadepath='haarcascade_frontalface_default.xml'
```

```python
# Mengkonversi keabuan
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.show()
```

```python
# Mengaktifkan Haarcascade
haarcascade=cv2.CascadeClassifier(cv2.data.haarcascades+haarcascadepath)
```

```python
# proses deteksi membuat persegi
def deteksi_wajah(gbr):    
    face_gbr=gbr.copy()
    # proses deteksi dengan multiscal
    face_detect = haarcascade.detectMultiScale(face_gbr,scaleFactor=1.2,minNeighbors=2) 
    
    # buat persegi untuk deteksi wajah
    for (x,y,w,h) in face_detect:
        cv2.rectangle(face_gbr,(x,y),(x+w,y+h),(0,255,255),5)
        # cv2.imshow(face_gbr)
    return face_gbr
```

```python
z=deteksi_wajah(image)
# tampil hasil deteksi
plt.imshow(z)
plt.show()
```
