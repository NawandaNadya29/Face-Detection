```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Menginput Gambar
#SMILEDETECTION
image = cv2.imread('blekpingsenyum.jpeg')

plt.imshow(image)
plt.show()
```

```python
# Mengkonversikeabuan

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.show()
```

```python
haarcascadepath='haarcascade_smile.xml'
haarcascade=cv2.CascadeClassifier(cv2.data.haarcascades+haarcascadepath)
```

```python
smile_detect = haarcascade.detectMultiScale(gray,scaleFactor=2.0, minNeighbors=15) 
```

```python
def deteksi_senyum(gbr):
  for senyum in smile_detect:
    x,y,w,h=senyum
    cv2.rectangle(image, (x,y), (x+w, y+h), (0,255,255),3)
  return image
```

```python
x=deteksi_senyum(gray)
plt.imshow(x)
plt.show()
```
